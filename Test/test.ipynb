```python
raw_data = pd.read_table('yesware-test-data-v1-6.txt', names=('A'))
```

```python
df = pd.DataFrame(raw_data)
df.columns=['original']
```

```python
df['split'] = df['original'].split('--')
```

```python
# below line parses the entire df, removes everything after the '--' & writes data to a new column called split
df['split'] = df['original'].map(lambda x: x.split('--', 1)[0])
```

```python
# below to show if the data in the column contains the specified character
# useing this column in the future to check if any non name rows get selected in new_df
df['required'] = df.split.str.contains(',') == True
```
